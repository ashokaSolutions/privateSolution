```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.preprocessing import LabelEncoder
data=pd.read_csv("D:\\python\\lab\\dataScience\\privateSolution\\DALastSem\\files\\Fish.csv")
print(data.head(5))
```

```python
print(data.info())
```

```python
sns.pairplot(data)
```

```python
x=data[["Height","Width","Length1","Length2","Length3"]].values
y=data["Weight"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
model=LinearRegression().fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
print(r2_score(y_test,y_pred))
# slope=x_train*model.coef_+model.intercept_#mx+c
# plt.scatter(x_train,y_train,color="blue")
# plt.plot(x_train,slope,color="red")
# plt.show()
plt.scatter(x_test[:,1], y_test, color='green', label = 'Actual Weight')
plt.scatter(x_test[:,1], model.predict(x_test), color='red', label = 'Prdicted Weight')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Linear Regression Model')
plt.show()

```
