```python
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plat
import numpy as np
import statsmodels #use function(ols summary:sumrray like r2)
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split #training keliye adha and testing keliye half data from total data
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error #model evaluation
data=pd.read_csv("C:\\Users\\ahire\\OneDrive\\Desktop\\newHouse.csv")#1
data.head()
```

```python
data=data.dropna()#2
print(data.info())#3
x=np.array(data[["TotalBsmtSF"]])#4.1
y=np.array(data[["SalePrice"]])#4.2
plt.scatter(x,y,color="r")#5.1
plt.show()#5.2
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)#6 imp
li=LinearRegression()
li.fit(x_train,y_train)#7
y_pred=li.predict(x_test)#8
```

```python
sns.regplot(x="TotalBsmtSF",y="SalePrice",data=data,color="red",line_kws={"color": "blue"})#9
print(li.coef_)
```

```python
test_accuracy=r2_score(y_test,y_pred)#10
print(test_accuracy)#11
```
