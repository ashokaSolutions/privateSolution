```python
from PIL import Image
import numpy as np #numerical
import pandas as pd #patterns
import seaborn as sns
import statsmodels #use function(ols summary:sumrray like r2)
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split #training keliye adha and testing keliye half data from total data
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error #model evaluation
#------------------------------data gathering-----------------------------
data=pd.read_csv("C:\\Users\\ahire\\OneDrive\\Desktop\\Advertising.csv")
data.head()
#-------------------------------EDA----------------------------------------
print("column:"+data.columns)
print("******shape******")
print(data.shape)
print("******size******")
print(data.size)
print("******info******")
print(data.info)
#=============drop column================
sns.pairplot(x_vars=["TV Ad Budget ($)","Radio Ad Budget ($)","Newspaper Ad Budget ($)"],y_vars="Sales ($)",data=data)
plt.show() #to print the difference
#=============head map===================
#negative=they did not have any relation ship(error points)
sns.heatmap(data.corr(),annot=True)
#tv and budge liya kiu ki they were tunnel shape(start se uper tak varibale thode spread out hote hai)
```

```python
#========================INDEPENDENT AND DEPENDENT VARIBALE SHOW========================
x=data[["TV Ad Budget ($)"]] #independent
y=data[["Sales ($)"]]        #dependent
#train and split:split "both" the variables x and y,0.3 means 30 split 70:30 karna tha
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
print(x_test)
li=LinearRegression()
li.fit(x_train,y_train)
#r=-1 perfect negative
#r=+1 prefect positive
#r=0 no correlation
li.coef_#c
y_pred=li.predict(x_test)
print(y_pred)
```

```python
#test the accuracy
test_accuracy=r2_score(y_test,y_pred)
print(test_accuracy)

#mean squared
print(mean_absolute_error(y_test,y_pred))
```
