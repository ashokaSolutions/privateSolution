```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import seaborn as sns
data=pd.read_csv("C:\\Users\\ahire\\OneDrive\\Desktop\\Advertising.csv")
data=data.drop(["Unnamed: 0"],axis=1)
print(data.head(5))
```

```python
sns.pairplot(data)
```

```python
#check in the section of dependent var and check for most accurat tunnel from that
x=data[["TV Ad Budget ($)"]]
y=data["Sales ($)"]
```

```python
#we split our data set and store it in x_train,y_train
#then we use to predict the other set and match it with x_test
#now when we use train_test_split(splits 4 subset)
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.3,random_state=1)
model=LinearRegression().fit(x_train,y_train)
y_pred=model.predict(x_test) #we predicted the values using x_test,unseen dataset we cal y_pred
test_1accuracy=r2_score(y_test,y_pred)#see how same they are
test_1accuracy
```

```python
"""
This code would work if you want to plot the regression line on the test data set. However
"""
plt.scatter(x_test,y_test,color="red")
plt.plot(x_test,y_pred,'b')#y=mx+c
plt.show()
```

```python
"""
This code would work if you want to plot the regression line on the test data set. However
"""
plt.scatter(x_train,y_train,color="red")
plt.plot(x_train,model.coef_*x_train+model.intercept_,'b')#y=mx+c
plt.show()
```
